```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
DATA_DIR = 'data'
MET_DOT_HU_DATA = os.path.join(DATA_DIR, 'BP_d.txt')
```

```python
data = pd.read_csv(MET_DOT_HU_DATA, sep=';')
data
```

```python
data['year'] =[int(a[0:4]) for a in data['#datum']]
data['month'] =[int(a[5:7]) for a in data['#datum']]
data['day'] =[int(a[8:10]) for a in data['#datum']]
```

```python
plt.figure(figsize=(20,10))
plt.plot(data['d_tx'])
plt.show()
```

```python
plt.figure(figsize=(20,10))
years = data['year'].unique()
year_avg = data.groupby(['year'])['d_tx'].mean()
poly_order = 1
params = np.polyfit(years, year_avg, poly_order)
year_avg_first_order_poly_fnc = np.poly1d(params)
plt.plot(years, np.stack([year_avg,  year_avg_first_order_poly_fnc(years)], 1))
plt.title('Yearly average temperature')
plt.show()
```

# Conclusion
The year parameter seems important, but it varries greatly ~4 degree between the smallest and highest average. It is possible that the model will overfit on it.

```python
year = np.array(data['year'])
month = np.array(data['month'])
day = np.array(data['day'])
d_tx = np.array(data['d_tx'])
```

```python
data_one_day = np.stack([year, month, day, d_tx], 1)
```

```python
memory_size = 7
data_many_day_X = np.array([data_one_day[i:i+memory_size, :].flatten() for i in range(0, data_one_day.shape[0] - memory_size)])
data_many_day_Y = np.array([data_one_day[i, 3].flatten() for i in range(memory_size, data_one_day.shape[0])])
```

```python
print(data_many_day_X.shape, data_many_day_Y.shape, data.shape)
print(data_many_day_X[-1, :], data_many_day_Y[-1], data.loc[40177-10:40177, 'd_tx'])
```

```python
print(data_many_day_X.shape, data_many_day_Y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

train_data_many_day_X, dev_test_data_many_day_X, train_data_many_day_Y, dev_test_data_many_day_Y = \
    train_test_split(data_many_day_X, data_many_day_Y, test_size=0.3, random_state=42)
dev_data_many_day_X, test_data_many_day_X, dev_data_many_day_Y, test_data_many_day_Y = \
    train_test_split(dev_test_data_many_day_X, dev_test_data_many_day_Y, test_size=0.66, random_state=42)
normalizerX = MinMaxScaler().fit(train_data_many_day_X)
normalizerY = MinMaxScaler().fit(train_data_many_day_Y)
train_data_many_day_X = normalizerX.transform(train_data_many_day_X)
dev_data_many_day_X = normalizerX.transform(dev_data_many_day_X)
test_data_many_day_X = normalizerX.transform(test_data_many_day_X)
train_data_many_day_Y = normalizerY.transform(train_data_many_day_Y)
dev_data_many_day_Y = normalizerY.transform(dev_data_many_day_Y)
test_data_many_day_Y = normalizerY.transform(test_data_many_day_Y)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=64, activation='relu', input_dim=4*7))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['accuracy'])
model.fit(train_data_many_day_X, train_data_many_day_Y, epochs=30, batch_size=32)
```

```python
model.summary()
```

```python
model.evaluate(dev_data_many_day_X, dev_data_many_day_Y)
model.evaluate(test_data_many_day_X, test_data_many_day_Y)
```

```python
normalizerY.inverse_transform(model.predict(test_data_many_day_X))
```
